```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
happy_image = cv2.imread('happyboy.jpg')
```

```python
plt.imshow(happy_image)
```

```python
plt.imshow(cv2.cvtColor(happy_image, cv2.COLOR_BGR2RGB))
```

```python
from deepface import DeepFace
```

```python
predictions = DeepFace.analyze(happy_image)
```

```python
print(predictions)
```

```python
print(predictions['dominant_emotion'])
```

```python
facecascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
gray=cv2.cvtColor(happy_image,cv2.COLOR_BGR2GRAY)
```

```python
faces=facecascade.detectMultiScale(gray,1.1,4)
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(happy_image,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
plt.imshow(cv2.cvtColor(happy_image,cv2.COLOR_BGR2RGB))
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(happy_image,predictions['dominant_emotion'],(0,50),font,3,(0,0,255),2,cv2.LINE_4)
```
